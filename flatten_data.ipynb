### Flatten the paragraph csv, keep “Topic”, “Prompt”, “Paragraph ID”, and “Sentence ID” 

```python
import pandas as pd
import io
import requests

raw_data=pd.read_csv("data/paragraph_raw.csv")
```

```python
data = raw_data.iloc[:,3:]
```

```python
raw_data[:3]
```

```python
flattened_sent = data.values.flatten()
flattened_sent[:3]
```

```python
flattened_sent = pd.DataFrame(flattened_sent)
flattened_sent.columns = ['sentence']
flattened_sent[:3]
```

```python
topic = list(raw_data["Topic"])
topics = []
for i in range(len(raw_data)):
    topics.append([topic[i]]*10)
topics[:3]
```

```python
prompt = list(raw_data["Prompt"])
prompts = []
for i in range(len(raw_data)):
    prompts.append([prompt[i]]*10)
prompts[:3]
```

```python
paragraph_id = list(raw_data["Paragraph ID"])
paragraph_ids = []
for i in range(len(raw_data)):
    paragraph_ids.append([paragraph_id[i]]*10)
paragraph_ids[:3]
```

```python
indices = []
for i in range(len(raw_data)):
    indices.append([i]*10)
indices[:3]
```

```python
import numpy as np

flattened_indices = np.array(indices).flatten()
print(flattened_indices[:21])

flattened_topics = np.array(topics).flatten()
print(flattened_topics[:21])

flattened_prompts = np.array(prompts).flatten()
print(flattened_prompts[:21])

flattened_paragraph_ids = np.array(paragraph_ids).flatten()
print(flattened_paragraph_ids[:21])
```

```python
# check if dimension the same
len(flattened_indices) == len(flattened_sent)
```

```python
flattened_sent['index'] = flattened_indices
flattened_sent['topic'] = flattened_topics
flattened_sent['prompt'] = flattened_prompts
flattened_sent['paragraph_id'] = flattened_paragraph_ids

flattened_sent[:21]
```

```python
# Get rid of nans
output = flattened_sent.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```

```python
output[:21]
```

```python
output.to_csv("data/flattened_paragraphs.csv")
```
